```python
y = data["Performance_Score"]
x = data[["Years_At_Company","Monthly_Salary","Overtime_Hours","Promotions","Employee_Satisfaction_Score"]]
```

```python
x.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size= 0.2)
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
import joblib
joblib.dump(scaler,"scaler.pkl")
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.metrics import accuracy_score
```

```python
def modelperformance(prediction):
  print("Accuracy score on model is {}".format(accuracy_score(y_test, predictions)))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
log_model.fit(X_train, y_train)
```

```python
y
```

```python
log_model.predict([[0,0,0,0,0]])
```

```python
log_model.predict(X_test)
```

```python
log_predictions = log_model.predict(X_test)
```

```python
def modelperformance(prediction):
  print("Accuracy score on model is {}".format(accuracy_score(y_test, prediction)))

modelperformance(log_predictions)
```

```python
from sklearn.neighbors import KNeighborsClassifier
param_grid = {"n_neighbors": [3,5,7,9,11],
              "weights":["uniform","distance"]}
```

```python
gridkn = GridSearchCV(KNeighborsClassifier(), param_grid, cv = 5)
```

```python
gridkn.fit(X_train, y_train)
```

```python
gridkn.best_params_
```

```python
predictions = gridkn.predict(X_test)
```

```python
modelperformance(predictions)
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV

svm = LinearSVC(random_state=42)

param_grid = {
    'C': [0.01, 0.1, 0.5, 1]
}

gridsvc = GridSearchCV(
    svm,
    param_grid,
    cv=3,
    n_jobs=-1
)

gridsvc.fit(X_train, y_train)
```

```python
X_train.shape
```

```python
param_grid
```

```python
y_pred = gridsvc.predict(X_test)
```

```python
modelperformance(y_pred)
```

```python
joblib.dump(gridsvc,"model.pkl")
```

```python
joblib.dump(scaler, "scaler.pkl")
joblib.dump(gridsvc, "model.pkl")
```

```python
import joblib
import numpy as np

scaler = joblib.load("scaler.pkl")
model = joblib.load("model.pkl")

years = float(input("Enter years at company: "))
salary = float(input("Enter monthly salary: "))
overtime = float(input("Enter overtime hours: "))
promotions = float(input("Enter number of promotions: "))
satisfaction = float(input("Enter satisfaction score: "))

X = np.array([[years, salary, overtime, promotions, satisfaction]])

X_scaled = scaler.transform(X)

prediction = model.predict(X_scaled)[0]

print("\nPredicted Performance Score:", prediction)
```
